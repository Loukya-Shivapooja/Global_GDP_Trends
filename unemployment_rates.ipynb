```python
#Import Depenencies

import pandas as pd
```

```python
#Creating a path to the .csv file.
file_to_unemployment = "Resources_1/unemployment_rates.csv"
file_to_codes = "Resources_1/country-codes.csv"
```

```python
unemp_rates_csv = pd.read_csv(file_to_unemployment)
```

```python
#Create a data frame with the unemployment rates csv.
unemp_rates_df = pd.DataFrame(unemp_rates_csv)
unemp_rates_df
```

```python
#Drop first 3 rows with no data
unemp_rates_df_clean = unemp_rates_df.iloc[3:, :]

#Use new first row as headers
headers = unemp_rates_df_clean.iloc[0]
unemp_rates_df_clean = unemp_rates_df_clean[1:]
unemp_rates_df_clean.columns = headers

unemp_rates_df_clean.head(10)
```

```python
#Drop the columns we don't need for our analysis.
unemploy_rates_df = unemp_rates_df_clean.drop(unemp_rates_df_clean.columns[2:45], inplace=False, axis=1)
unemploy_rates_df
```

```python
unemploy_rates_df.drop(unemploy_rates_df.columns[21:23], inplace=True, axis=1)
```

```python
unemploy_rates_df
```

```python
#Reformat the column names.
unemploy_rates_df = unemploy_rates_df.set_axis(['Country Name', 'Country Code', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008',\
                                   '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019'], axis=1)
unemploy_rates_df
```

```python
#Identify any null values in our dataset.
unemploy_rates_df.count()
```

```python
#Drop the null values in our dataset.
unemploy_rates_df.dropna(inplace=True)
```

```python
unemploy_rates_df
```

```python
#Add the country codes csv
country_codes_df = pd.read_csv(file_to_codes)
country_codes_df
```

```python
#Drop columns from the country code csv
country_codes_df.drop(country_codes_df.columns[3:5], inplace=True, axis=1)
```

```python
#Drop the Alpha-2 code.
country_codes_df.drop(country_codes_df.columns[1], inplace=True, axis=1)
```

```python
#Rename the Alpha-3 code
country_codes_df = country_codes_df.rename(columns={"Alpha-3 code": "Country Code","English short name lower case":"Country Name"})
```

```python
country_codes_df
```

```python
countries_unemployment_df = pd.merge(unemploy_rates_df, country_codes_df, on= "Country Code", suffixes=["","_codes"])
countries_unemployment_df
```

```python
countries_unemployment_df.count()
```

```python
#Remove country codes_names column
countries_unemployment_df.drop(columns="Country Name_codes", inplace=True)
```

```python
countries_unemployment_df
```

```python
#Import the countries income level to merge with the rest of our dataset.
file_income = pd.read_csv("Resources_1/countries_region_income.csv")
```

```python
#Create a Dataframe for the countries income csv.
income_df = pd.DataFrame(file_income)
income_df
```

```python
#Drop columns we don't need.
income_df.drop(["SpecialNotes", "Region", "TableName"], axis=1, inplace=True)
```

```python
income_df
```

```python
final_unemployment_df = pd.merge(countries_unemployment_df, income_df, on="Country Code")
final_unemployment_df
```

```python
#Check for any null values.
final_unemployment_df.count()
```

```python
#Format the percentages to the nearest tenths place.
final_unemployment_df.round(2)
```

```python
#Check the data types.
final_unemployment_df.dtypes
```

```python
#Export final dataset to a csv file without an index.
final_unemployment_df.to_csv("unemployment_rates_data.csv", index=False)
```

```python
#Find the top ten countries in 2019.
top_countries = final_unemployment_df.sort_values(by="2019", ascending=False).round(2)
top_countries.head(10)
```

```python
#Find the bottom ten countries in 2019.
bottom_countries = final_unemployment_df.sort_values(by="2019", ascending=True).round(2)
bottom_countries.head(10)
```

```python
#Find the average for each year.
average_unemployment = final_unemployment_df.mean(numeric_only=True).round(2)
```

```python
average_unemployment
```

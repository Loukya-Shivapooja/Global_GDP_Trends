```python
#Import Depenencies

import pandas as pd
```

```python
#Creating a path to the .csv file.

file_to_path = "military_expenditure.csv"
```

```python
mil_csv = pd.read_csv(file_to_path)
```

```python
#Creating a DataFrame using Pandas

mil_df = pd.DataFrame(mil_csv)
mil_df
```

```python
#Drop first 3 rows with no data
mil_df_clean = mil_df.iloc[3:, :]

#Use new first row as headers
headers = mil_df_clean.iloc[0]
mil_df_clean = mil_df_clean[1:]
mil_df_clean.columns = headers

mil_df_clean.head(10)
```

```python
mil_exp_df = mil_df_clean.drop(mil_df_clean.columns[2:45], inplace=False, axis=1)
mil_exp_df
```

```python
mil_exp_df.drop(mil_exp_df.columns[22], inplace=True, axis=1)
```

```python
mil_exp_df = mil_exp_df.set_axis(['Country Name', 'Country Code', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008',\
                                   '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020'], axis=1)
mil_exp_df
```

```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Import Data file
health_exp = pd.read_csv("Resources/health_expenditure.csv")
health_exp_df = pd.DataFrame(health_exp)
health_exp_df
```

```python
# Droping the top 3 rows
health_exp_df = health_exp_df.drop(range(0,3))
health_exp_df
```

```python
# Setting first column as Index
new_header = health_exp_df.iloc[0]
health_exp_df = health_exp_df[1:]
health_exp_df.columns = new_header
health_exp_df
```

```python
# Name of the columns
health_exp_df.columns
```

```python
# Type of columns
health_exp_df.dtypes
```

```python
# Dropping the columns from year 1960 to 2000
health_exp_df = health_exp_df.drop(health_exp_df.columns[2:45], inplace = False, axis = 1)
health_exp_df
```

```python
# Dropping the 2021 column
health_exp_df = health_exp_df.drop(columns = health_exp_df.columns[21:22], inplace = False, axis = 1)
health_exp_df
```

```python
health_exp_df.columns
```

```python
# Setting the Index
health_exp_df = health_exp_df.set_axis(['Country Name', 'Country Code', '2001', '2002', '2003', '2004', '2005',         
                                        '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', 
                                        '2015', '2016', '2017', '2018', '2019'], axis = 1)
health_exp_df
```

```python
health_exp_df.count()
```

```python
health_exp_df = health_exp_df.dropna()
health_exp_df
```

```python
# Resetting the index to Zero
health_exp_df = health_exp_df.reset_index(drop=True)
health_exp_df
```

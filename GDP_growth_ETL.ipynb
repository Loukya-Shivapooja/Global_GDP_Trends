```python
#initial Imports
import pandas as pd
import os
```

```python
#csv files toload
file_path = ("Resources/GDP_growth.csv")
country_code_file_path = ("Resources/country-codes.csv")
countries_region_income = ("Resources/countries_region_income.csv")
```

## Clean GDP_Growth

```python
#read in df
GDP_growth = pd.read_csv(file_path)
df = pd.DataFrame(GDP_growth)
df.head(10)
```

```python
#get header rows
headers = []
```

```python
#drop first 3 rows with no data
GDP_growth_df = df.iloc[3:, :]

#use new first row as headers
headers = GDP_growth_df.iloc[0]
GDP_growth_df = GDP_growth_df[1:]
GDP_growth_df.columns = headers

GDP_growth_df.head(10)
```

```python
#drop columns that will not be of use
GDP_growth_df = GDP_growth_df.drop(GDP_growth_df.columns[2:45], inplace=False, axis=1)
GDP_growth_df = GDP_growth_df.drop(GDP_growth_df.columns[21:23], inplace=False, axis=1)
GDP_growth_df
```

```python
#clean Up DataFrame
#replace column names
GDP_growth_df = GDP_growth_df.set_axis(['Country Name', 'Country Code', '2001', '2002','2003','2004','2005','2006',
                                            '2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017',
                                            '2018','2019'], axis = 1)
GDP_growth_df
```

```python
#count null values in each row
GDP_growth_df.isnull().sum()
```

```python
#drop Null Values
GDP_growth_df = GDP_growth_df.dropna()
GDP_growth_df
```

```python
#check null values
GDP_growth_df.isnull().sum()
```

```python
#check the number of rows
GDP_growth_df.count()
```

```python
#check country names for strings that are not countries
GDP_growth_df['Country Name'].unique().tolist()
```
